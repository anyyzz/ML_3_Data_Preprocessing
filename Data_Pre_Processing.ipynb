## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan , strategy='mean') # object= imputer, class= SimpleImputer
imputer.fit(X[:,1:3]) # fit method looks for all missing values
X[:,1:3]=imputer.transform(X[:,1:3]) # transform method fills in the missing values
```
